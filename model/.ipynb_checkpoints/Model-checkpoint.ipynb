```python
import tensorflow as tf
import pandas as pd
import numpy as np
```

```python
features = pd.read_csv('./spectrograms.csv')
```

```python
x_data = features.values[:, :121].reshape((-1,11,11,1))
```

```python
x_data = (x_data - np.min(x_data))/(np.max(x_data)- np.min(x_data))
```

```python
x_data.shape
```

```python
n_epochs = 300
batch_size = 60
learning_rate = 0.001
x_len = x_data.shape[0]
```

```python
tf.reset_default_graph()
```

```python
with tf.variable_scope('input'):
    X = tf.placeholder(dtype=tf.float32, name='X', shape=[None,11,11,1])
    y = tf.placeholder(dtype=tf.float32, name='y', shape=[None,11,11,1])
    
with tf.variable_scope('stackedAE'):
    layer1 = tf.layers.conv2d(activation=tf.nn.elu, filters=2, inputs=X, kernel_initializer=tf.contrib.layers.variance_scaling_initializer(),name='layer1', padding='SAME', strides=[1,1], kernel_size=[4,4], kernel_regularizer=tf.contrib.layers.l1_regularizer)
    layer2 = tf.layers.conv2d_transpose(activation=None, filters=1, inputs=layer1, kernel_initializer=tf.contrib.layers.variance_scaling_initializer(), name='layer2', padding='SAME', strides=[1,1], kernel_size=[4,4])

with tf.variable_scope('cost'):
    l1_loss = tf.get_collection(tf.GraphKeys.REGULARIZATION_LOSSES)
    loss = tf.reduce_mean(tf.reduce_sum(tf.nn.sigmoid_cross_entropy_with_logits(logits=layer2, labels=y), axis=1))
    optimizer = tf.train.MomentumOptimizer(momentum=0.99, learning_rate=learning_rate,use_nesterov=True, name='optimizer')

with tf.variable_scope('training_op'):
    training_op = optimizer.minimize(loss)
```

```python
x_train, x_test = x_data[:170], x_data[170:]
x_train.shape, x_test.shape
```

```python
def next_batch(x, batch_size):
    ix = np.random.permutation(x.shape[0])[:batch_size]
    return x[ix]
```

```python
init = tf.global_variables_initializer()
with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for iteration in range(x_len//batch_size):
            x_batch = next_batch(x_train, batch_size)
            sess.run(training_op, feed_dict = {X:x_batch, y:x_batch})
        mse_train = loss.eval(feed_dict={X:x_train, y:x_train})
        mse_test = loss.eval(feed_dict = {X:x_test, y:x_test})
        print('epoch: %d, train loss: %d, test loss: %d'%(epoch, mse_train, mse_test))
```
